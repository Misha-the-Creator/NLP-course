```python
import json
import translators as ts
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
with open('example_texts.json', 'r') as json_file:
    texts = json.load(json_file)
```

```python
class Text_Formatter():
    def __init__(self, text, replace_dict):
        self.text = text
        self.replace_dict = replace_dict
    
    def symbol_replace(self):
        result = []
        for text in self.text:
            for old, new in self.replace_dict.items():
                text = text.replace(old, new)
            result.append(text)
        return result

replace_dict = {'\n\n': '\n',
                ' ,': ',',
                ' .': '.',
                '( ': '(',
                ' )': ')',
                ' - ': '-',
                '&quot; ': '"',
                ' &quot;': '"',
                ' :': ':',
                '¬´ ': '"',
                ' ¬ª': '"'}

formatter = Text_Formatter(texts, replace_dict)
formatted = formatter.symbol_replace()

print(formatted)
```

```python
for text in formatted:
    print('==='*200)
    print('–ò—Å—Ö–æ–¥–Ω—ã–π —Ç–µ–∫—Å—Ç¬†üìÉ:')
    print(text)
    print(' ')
    print(f"–ï–≥–æ –ø–µ—Ä–µ–≤–æ–¥ {'üá¨üáß'}")
    print(f"{ts.translate_text(text, 'yandex', to_language='en')}")
    translated_text = ts.translate_text(text, 'yandex', to_language='en')
    print(' ')
    print(f"–ï–≥–æ —Å—É–º–º–∞—Ä–∏–∑–∞—Ü–∏—è {'‚úçÔ∏è'}")
    summary = summarizer(translated_text, max_length=250, min_length=100, do_sample=False)
    print(summary[0]['summary_text'])
```
